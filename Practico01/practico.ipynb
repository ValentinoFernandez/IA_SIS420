```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

datos = pd.read_csv('datos_peso_estatura.csv')

plt.scatter(datos['Estatura'], datos['Peso'])
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.title('Distribución de Peso vs Estatura')
plt.show()

def modelo(x, a, b, c):
    return a * x**2 + b * x + c

x_data = datos['Estatura']
y_data = datos['Peso']

popt, pcov = curve_fit(modelo, x_data, y_data)
a, b, c = popt

print(f'Parámetros del modelo: a={a}, b={b}, c={c}')

x_line = np.linspace(min(x_data), max(x_data), 100)
y_line = modelo(x_line, *popt)

plt.scatter(x_data, y_data, label='Datos')
plt.plot(x_line, y_line, color='red', label='Curva ajustada')
plt.xlabel('Estatura (m)')
plt.ylabel('Peso (kg)')
plt.title('Curva ajustada de Peso vs Estatura')
plt.legend()
plt.show()
```
